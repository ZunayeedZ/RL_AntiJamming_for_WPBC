```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pyl
from math import pi
fm = eval(input('Enter the frequency of the sinusoid: '))
fs = eval(input('Enter the sampling frequency: ')) #fs >= 2fm (Sampling theorem)
t = np.arange(0,0.05,0.0005) #min,max,step size
t1 = np.arange(0,0.05,(1/fs))
x = np.cos(2*pi*fm*t)
x1 = np.cos(2*pi*fm*t1)
plt.plot(t,x)
plt.stem(t1,x1,'r*')
plt.show
plt.grid()
```

In above case fs = 10*fm (sampling freq 10 times signal freq)

Let's try combining 2 different signals

```python
Ac = 1 #amplitude of the incident RF signal from AP
Am = 1 #amplitude of the backscattered signal from BD
#fc = 2.46*10**9 #2.462GHz, reference: "HitchHike: Practical Backscatter Using Commodity WiFi"
#fm = 50*10**6 #50MHz, reference: "HitchHike: Practical Backscatter Using Commodity WiFi"
fc = 100
fm = 5
t = np.arange(-0.25,0.25,1*10**-3) #min,max,step size
c = Ac*np.cos(2*pi*fc*t)
m = Am*np.cos(2*pi*fm*t)
S = (c)*(m) #carrier multiplied with the message
plt.plot(t,S)
#plt.stem(t1,x1,'r*')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Modulated signal')
plt.show
plt.grid()
print(np.shape(t),np.shape(S))
```
